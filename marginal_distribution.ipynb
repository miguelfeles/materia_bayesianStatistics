**Probability Mass Function (PMF)**
- **Type of Variable:** Discrete
- **Notation:** `p(X=x)` or `p(x)`
- **Definition:** The PMF gives the probability that a discrete random variable `X` is exactly equal to some value `x`.
- **Properties:**
  - `0 <= p(x) <= 1` for any value of `x`.
  - The sum of all possible values of `p(x)` equals 1, i.e., `Σ p(x) = 1`.
- **Usage:** PMF is used to describe the distribution of discrete random variables, like the outcome of rolling a die or the number of heads in coin flips.


A **discrete marginal distribution** refers to the probability distribution of a subset of discrete random variables, regardless of the values of the other variables in the distribution. It is obtained by summing (or aggregating) the probabilities over the values of the variables that are not in the subset. This process effectively "marginalizes out" the variables not of interest, leaving the distribution of the variable(s) in focus.

This starts when given a joint probability mass function (PMF) `p(X, Y)` for discrete random variables `X` and `Y`, the marginal distribution of `X` is calculated as:

p(X=x) = Σ p(X=x, Y=y)


where the summation is over all possible values of `Y`. This yields the distribution of `X` independent of `Y`.


````python
import numpy as np
import matplotlib.pyplot as plt

# Simulate rolling two dice
num_rolls = 10000
die1_rolls = np.random.randint(1, 7, num_rolls)
die2_rolls = np.random.randint(1, 7, num_rolls)

# Calculate marginal distribution of die1
values, counts = np.unique(die1_rolls, return_counts=True)
marginal_distribution_die1 = counts / num_rolls

# Plotting
plt.bar(values, marginal_distribution_die1)
plt.xlabel('Die Value')
plt.ylabel('Probability')
plt.title('Marginal Distribution of Die 1 Rolls')
plt.xticks(values)
plt.show()
````

A **continuous marginal distribution** is the probability distribution of a subset of continuous random variables within a joint distribution, disregarding the values of the other variables. It is derived by integrating the joint probability density function (PDF) over the range of the variables that are not of interest, effectively "marginalizing out" these variables.

For continuous random variables `X` and `Y` with a joint PDF `f(X, Y)`, the marginal distribution of `X` can be found by:

f(X=x) = ∫ f(X=x, Y=y) dy


This integral is taken over all possible values of `Y`, resulting in the marginal PDF of `X`, which describes the distribution of `X` irrespective of `Y`.

**Probability Density Function (PDF)**
- **Type of Variable:** Continuous
- **Notation:** `f(X=x)` or `f(x)`
- **Definition:** The PDF is used to specify the probability of a continuous random variable falling within a particular range of values. It represents the density of probability rather than the probability itself.
- **Properties:**
  - `f(x) >= 0` for all `x`.
  - The integral of `f(x)` over the entire space is 1, i.e., `∫ f(x) dx = 1`.
- **Usage:** PDF is utilized for continuous random variables, such as the measurement of height, weight, or temperature.

**Joint Probability Density Function (Joint PDF)**
- **Type of Variables:** Continuous
- **Notation:** `f(X=x, Y=y)` or `f(x, y)`
- **Definition:** The joint PDF describes the probability density of two continuous random variables, `X` and `Y`, occurring together. It gives the density of probability at any point `(x, y)` in the space of `X` and `Y`.
- **Properties:**
  - `f(x, y) >= 0` for all `(x, y)`.
  - The double integral of `f(x, y)` over all possible values of `x` and `y` equals 1, i.e., `∫∫ f(x, y) dx dy = 1`.
- **Usage:** Joint PDF is critical for analyzing the relationship and probability distribution of multiple continuous variables together.


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, multivariate_normal

# Parameters for the bivariate normal distribution
mu_x = 170  # mean height in cm
sigma_x = 10  # standard deviation of height
mu_y = 70  # mean weight in kg
sigma_y = 15  # standard deviation of weight

# Generate samples from a bivariate normal distribution
num_samples = 10000
covariance_matrix = np.diag([sigma_x**2, sigma_y**2])  # Assuming independence
mean_vector = [mu_x, mu_y]
samples = np.random.multivariate_normal(mean_vector, covariance_matrix, size=num_samples)

# Extracting heights (X values)
heights = samples[:, 0]

# Calculate marginal distribution of X (heights) directly
# Since we know it's normal, we can use the parameters directly
x_values = np.linspace(mu_x - 3*sigma_x, mu_x + 3*sigma_x, 1000)
marginal_pdf_x = norm.pdf(x_values, mu_x, sigma_x)

# Plotting
plt.plot(x_values, marginal_pdf_x)
plt.xlabel('Height (cm)')
plt.ylabel('Density')
plt.title('Marginal Distribution of Height')
plt.show()
````

# Independence of Random Variables

Independence between random variables implies that the occurrence of an event for one variable does not affect the probability of an event for another variable. This concept is applicable to both discrete and continuous random variables.

## Discrete Random Variables

For discrete variables `X` and `Y`, they are independent if:

```math
P(X=x \text{ and } Y=y) = P(X=x) \times P(Y=y)


This concept is rooted in the Central Limit Theorem (CLT), a fundamental principle in statistics that suggests when an output is derived from the sum or average of a large number of largely independent factors, it is often reasonable to assume that the distribution of the output approximates a normal distribution. This holds true regardless of the original distributions of these factors, provided there are enough of them and they meet certain conditions. The beauty of the CLT lies in its ability to simplify complex distributions into a normal distribution, which is well-understood and easier to work with. This assumption is particularly useful in various fields of study, such as economics, psychology, and natural sciences, where it facilitates the application of statistical methods to analyze data and make inferences.
